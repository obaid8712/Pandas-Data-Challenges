## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find number of unique items
unique_count=purchase_data["Item ID"].unique()
num_unique=len(unique_count)
#find Average price
average_price=purchase_data["Price"].mean()
# Number of Purchases
num_purchases=purchase_data["Item Name"].count()
#Total Revenue
total_rev=purchase_data["Price"].sum()
#print(total_rev)
purchase_df=pd.DataFrame({
    "Number of Unique Items":[num_unique],
    "Average Price":['{:,.2f}'.format(average_price)],
    "Number of Purchases":[num_purchases],
    "Total Revenue":['${:,.2f}'.format(total_rev)]
})
purchase_df.head()
```

```python
#print(total_Player)
gender_df=purchase_data[["SN","Gender"]]
total_Player=gender_df["SN"].nunique()


#print(gender_df.groupby('SN')['Gender'].nunique())
df=purchase_data[["SN","Gender"]]
a = df[df.SN.notnull()].groupby('SN')['Gender'].unique()

gen_df=pd.DataFrame.from_records(a.values.tolist()).stack().value_counts()

gender_percent=(gen_df/total_Player)*100

pd.options.display.float_format = '{:,.2f}'.format
gender_df=pd.DataFrame({
        "Total Count":gen_df,
        "Percentage of Players":gender_percent
}, index=["Male","Female","Other / Non-Disclosed"])
#pd.reset_option=display.float_format
gender_df.head()
```
